## Strands for Phase 2, release 0.01 (2.01)

780 strands in total

This release is similar to 1.04. It has both incubations and no incubations and recovered and non-recovered types of strands. Each strand has fixed strength of 0.16 and varying radius values from 10 to 100. 

```julia
strengthArray = [0.16]
radiusArray = [10, 15, 20, 25, 30, 35, 40, 50, 60, 70, 80, 90, 100]
numRepeats = 10;
```

```julia
function createRepeats(parentStrand::Dict{Symbol,String})
    strands = Array{Dict{Symbol,String}}(undef,0)
    for s in strengthArray, r in radiusArray, n in 1:numRepeats
        strand = deepcopy(parentStrand)
        strand[:name] = strand[:name]*"-($s,$r)-repeat$n"
        strand[:strength] = "$s"
        strand[:radius] = "$r"
        push!(strands,strand)
    end
    strands
end;
```

```julia
baseStrand = Dict{Symbol,String}()
baseStrand[:seedTime] = "2021-07-22T21:00:00Z" #9am on Friday morning July 23 (in Auckland)
baseStrand[:endTime] = "2021-11-04T20:00:00Z"#9pm in the evening of Nov 4 (in Auckland)
baseStrand[:minimumAppVersion] = "59";
```

```julia
### The base strand for the SI epidemics
SIstrand = deepcopy(baseStrand)
SIstrand[:name] = "2.01-SI"
SIstrand[:seedingProbability] = "0.2"
SIstrand[:incubationMean] = "172800" #default for two days incubation
SIstrand[:incubationShape] = "3"
SIstrand[:infectionMean] = "31536000" #365 days
SIstrand[:infectiousPeriodShape] = "10000";
```

```julia
### SEI
SIstrandYesIncubation = deepcopy(SIstrand)
SIstrandYesIncubation[:name] = SIstrandYesIncubation[:name]*"-incYes"
SIstrandYesIncubation
```

```julia
##SI
SIstrandNoIncubation = deepcopy(SIstrand)
SIstrandNoIncubation[:name] = SIstrandNoIncubation[:name]*"-incNo"
SIstrandNoIncubation[:incubationMean] = "1" #1 second incubation
SIstrandNoIncubation[:incubationShape] = "10000"
SIstrandNoIncubation
```

```julia
#The base strand for the SIR epidemics
SIRstrand = deepcopy(baseStrand)
SIRstrand[:name] = "2.01-SIR"
SIRstrand[:seedingProbability] = "0.2" #will be this or 0.4
SIRstrand[:incubationMean] = "172800" 
SIRstrand[:incubationShape] = "3"
SIRstrand[:infectionMean] = "432000" #5 days should vary
SIRstrand[:infectiousPeriodShape] = "3";
```

```julia
### SEIR base strand
SIRstrandYesIncubation = deepcopy(SIRstrand)
SIRstrandYesIncubation[:name] = SIRstrandYesIncubation[:name]*"-incYes"
SIRstrandYesIncubation
```

```julia
##SIR base strand
SIRstrandNoIncubation = deepcopy(SIRstrand)
SIRstrandNoIncubation[:name] = SIRstrandNoIncubation[:name]*"-incNo"
SIRstrandNoIncubation[:incubationMean] = "1" #1 second incubation
SIRstrandNoIncubation[:incubationShape] = "10000"
SIRstrandNoIncubation
```

```julia
SIRstrandYesIncubationShortDuration = deepcopy(SIRstrandYesIncubation)
SIRstrandYesIncubationShortDuration[:name] = SIRstrandYesIncubationShortDuration[:name]*"-days5";
SIRstrandYesIncubationShortDuration
```

```julia
SIRstrandYesIncubationLongDuration = deepcopy(SIRstrandYesIncubation)
SIRstrandYesIncubationLongDuration[:infectionMean] = "864000" 
SIRstrandYesIncubationLongDuration[:name] = SIRstrandYesIncubationLongDuration[:name]*"-days10";
SIRstrandYesIncubationLongDuration
```

```julia
SIRstrandNoIncubationShortDuration = deepcopy(SIRstrandNoIncubation)
SIRstrandNoIncubationShortDuration[:name] = SIRstrandNoIncubationShortDuration[:name]*"-days5";
SIRstrandNoIncubationShortDuration
```

```julia
SIRstrandNoIncubationLongDuration = deepcopy(SIRstrandNoIncubation)
SIRstrandNoIncubationLongDuration[:infectionMean] = "864000" 
SIRstrandNoIncubationLongDuration[:name] = SIRstrandNoIncubationLongDuration[:name]*"-days10";
SIRstrandNoIncubationLongDuration
```

```julia
#An array that will hold all strands
strands = Array{Dict{Symbol,String}}(undef,0)
append!(strands, createRepeats(SIstrandYesIncubation))
append!(strands, createRepeats(SIstrandNoIncubation))
append!(strands, createRepeats(SIRstrandYesIncubationShortDuration))
append!(strands, createRepeats(SIRstrandYesIncubationLongDuration))
append!(strands, createRepeats(SIRstrandNoIncubationShortDuration))
append!(strands, createRepeats(SIRstrandNoIncubationLongDuration));
```

```julia
function strandJSON(sd::Dict{Symbol,String})
    str = """
        {\"name\":\"$(sd[:name])\",
         \"seedingProbability\":\"$(sd[:seedingProbability])\",
         \"infectionProbabilityMapP\":\"$(sd[:strength])\",
         \"infectionProbabilityMapK\":\"$(sd[:radius])\",
         \"infectionProbabilityMapL\":\"1000000\",
         \"incubationPeriodMeanSec\":\"$(sd[:incubationMean])\",
         \"incubationPeriodShape\":\"$(sd[:incubationShape])\",
         \"infectiousPeriodMeanSec\":\"$(sd[:infectionMean])\",
         \"infectiousPeriodShape\":\"$(sd[:infectiousPeriodShape])\",
         \"startTime\":\"$(sd[:seedTime])\",
         \"endTime\":\"$(sd[:endTime])\",
         \"minimumAppVersion\":\"$(sd[:minimumAppVersion])\"}
    """
    join(map(x -> isspace(str[x]) ? "" : str[x], 1:length(str))) #how the $%@$% to remove white space in Julia strings?
end

strandStrings = strandJSON.(strands)
if length(filter(x->x=="strands2_01.txt",map(abspath, readdir())))!=0
    rm("strands2_01.txt") #remove the text file in the working directory if exists
end
touch("strands2_01.txt") #create the text file in the working directory
open("strands2_01.txt","a") do file 
    newID = 565 #last strand in system
    for s in strandStrings
        newID +=1
        write(file,"$newID: $s\n")
        println("$newID: $s\n")
    end
end
```
